```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
apple_ticker=yf.Ticker('BTC-USD')
data = apple_ticker.history(start='2000-01-01', end = '2024-01-01')
```

```python

data['SMA50'] = data['Close'].rolling(window=10).mean()
data.head(55)
```

```python
data[['Close','SMA50']].plot(figsize=(10,5), title='Apple Price and 50 day simple moving average')
plt.show()
```

```python
data['SMA200'] = data['Close'].rolling(window=300).mean()
data['Signal'] = np.where(data['SMA50']>data['SMA200'],1,0) 
data.head(400)
```

```python
data['Market Return'] = data['Close'].pct_change()
data['Strategy Return']= data['Signal'].shift(1)*data['Market Return']
data['Cumulative_Market'] = (1 + data['Market Return']).cumprod()
data['Cumulative_Strategy'] = (1 + data['Strategy Return']).cumprod()
data[['Cumulative_Market', 'Cumulative_Strategy']].plot(figsize=(10, 5), title='Strategy vs. Buy & Hold ($1 Investment)')
plt.show()
```

```python
clean_data = data.dropna()
sharpe_ratio = clean_data['Strategy Return'].mean()/clean_data['Strategy Return'].std()*252**0.5
sharpe_ratio1 = (clean_data['Market Return'].mean()/clean_data['Market Return'].std())*252**0.5

print(sharpe_ratio1)
print(sharpe_ratio)

```
